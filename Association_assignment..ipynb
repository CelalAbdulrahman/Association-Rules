Read CSV data about Titanic passengers into a DataFrame

```python
titanic = pd.read_csv("Titanic.csv",header=None)
```

#Data Preprocessing

###One-hot encoding changes categories into binary format.

#so,

 That algorithms can work with it.

```python
df=pd.get_dummies(titanic)
df.head()
```

#Use Apriori Algorithm

###Change categories to binary format so that the algorithm can work on them.


Find frequent itemsets with 20% support using Apriori.

```python
frequents_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequents_itemsets
```

Use the lift metric to make association rules between traits.

```python
rules = association_rules(frequents_itemsets, metric="lift", min_threshold=0.6)
rules
```

#Association rules

###Get rules with a lot of confidence that show strong relationships.

```python
rules = association_rules(frequents_itemsets, metric="confidence", min_threshold=0.6)
rules
```

*Only take out rules that have a high level of confidence (60% or higher) and show strong.

```python
rules.sort_values('lift',ascending = False)[0:20]
```

 reliable connections between passenger traits*

```python
rules[rules['lift'] > 1]
```

#1.	What is lift and why is it important in Association rules?
#  Answer:
  A simple way to think about it is to picture yourself as a store owner.Lift lets you know if two products are really related or if they are just popular on their own.

 **Technical Definition:**
  Lift tells us how much more often two things are bought together than we would expect if they had nothing to do with each other.

 **For example:**
  If people buy bread 30% of the time.
  And 20% of the time, people buy butter.
  They should be bought together by chance 6% of the time (30% × 20%).
  But they are actually bought together 15% of the time.
  Then Lift = 15% ÷ 6% = 2.5

  **Why It's Important:**
  (i) Lift = 1: No connection, just random
  (ii) Lift > 1: There is a positive relationship between them (they draw each other in).
  (iii) Lift < 1: Bad relationship (they stay away from each other)

#2.	What is support and Confidence. How do you calculate them?
#  Answer:
  **Grocery Store Example:**
   Support = Score for Popularity
   Support(Bread) = How often bread is in all transactions = (Number of transactions with bread) ÷ (Total transactions) *For example, if bread is in 40 out of 100 bills, then support = 40%*

  **If-Then Reliability for Confidence:**
   Confidence (Bread → Butter) = How often do people who buy bread also buy butter? = (Transactions with both bread AND butter) ÷ (Transactions with bread only) *Example: 20 people bought both bread & butter out of 40 who bought bread → Confidence = 50%*

  **Math Formulas:**
  text Support(A) = P(A) = Count(A) / Total transactions
  Support(A → B) = P(A ∩ B) = Count(A & B together) / Total transactions
  Confidence(A → B) = P(B|A) = Support(A & B) / Support(A)

#3.	What are some limitations or challenges of Association rules mining?
# Answer
Big Problems:

** 1.The "Too Many Rules" Problem:**
   You get thousands of rules when you have a lot of things.
   Most of them are clear or not useful.
   "People who buy laptops also buy chargers" is an example.  Of course!

** 2.False Positives and Unreliable Correlations:**
   Sometimes finds patterns that are just random events
   "People who buy diapers also buy beer" is a well-known example that may only be true in certain situations.

** 3.Problems with the quality of the data:**
   Needs data that is clean and based on transactions
   Not having enough data can lead to wrong patterns.

** 4.Headache from Setting the Threshold:**
   Set support too high and you'll miss interesting, rare patterns.
   Set support too low, and you'll have too many rules to follow.
   It's like Goldilocks—you need values that are "just right."

** 5.Doesn't take the value of the item into account:**
   A rule about ₹10 chips is just as important as a ₹50,000 diamond.
   There is no value in business.

 **6.No cause and effect:**
   Just because A and B are bought together doesn't mean A makes B happen.
   Could be a third factor, like the festival season, that is causing both.

 **7. Memory and Computation:**
   The apriori algorithm can be slow when working with big datasets.
   Needs to be put into action quickly
